```python
import pandas as pd
```

```python
df = pd.read_csv('adult.data.csv')
df.head()
```

```python
pop_per_country = df.groupby(by='native-country').size()
```

```python
high_salary_df = df[df['salary'] == '>50K']
high_salary_count = high_salary_df.groupby('native-country').size()
# high_salary_count
```

```python
percentage_high_salary = (high_salary_count / pop_per_country) * 100
percentage_high_salary
```
